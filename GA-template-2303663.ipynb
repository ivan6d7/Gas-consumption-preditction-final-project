### <a id='toc1_5_4_'></a>[Task 4: Trends in average daily consumption](#toc0_)

The zoomed-in plots from Task 3 show short-term variations over a small part of the year, allowing you to spot peaks, drops, and unusual behavior within that period. However, this approach is limited to only that small part of the year. On the other hand, the full-year plots in Task 2 include all data points, covering the whole year, but are too dense to interpret clearly.

To balance detail with readability, while considering the whole year, the hourly data can be aggregated into **daily averages**. This smooths short-term fluctuations and makes it easier to compare households and to detect long-term or seasonal trends.

We use daily averages and not daily totals because some days may have *missing* hourly *data*. If we calculated daily totals, those days would appear to have lower consumption simply because some data points are missing, not because the household actually used less energy.
Using daily averages corrects for this: it represents the *typical hourly consumption* for that day, making comparisons between days and between households more meaningful and fair.

First, **create two new DataFrames** `df_household_1_agg` (for household 1) and `df_household_2_agg` (for household 2) using the functions `groupby` and `agg` (see the exercise EDA 11 for details) containing the following columns:

- `date` - the day of the observation
- `gas` - daily average gas consumption (mean of the hourly values for that day, in $\text{m}^3$)
- `electricity` - daily average electricity consumption (mean of the hourly values for that day, in $Wh$).

### <a id='toc1_5_9_'></a>[Replacing Gas with Electricity: Heat Pump Scenario](#toc0_)

So far, you have explored gas and electricity usage separately. Now, consider a hypothetical scenario: what if households replaced gas with electricity using modern heat pumps? 

Heat pumps are more efficient than direct electric heating. They achieve this efficiency by *moving heat from the environment into the building* rather than generating it directly. The Coefficient of Performance (COP) indicates how efficiently this transfer occurs. For modern heat pumps, we assume $\text{COP} = 4$, meaning the heat pump *delivers 4 kWh of heat for every 1 kWh of electricity it consumes*.

1 $m^3$ of natural gas (supplied in the Netherlands) contains about 9769 Wh of energy. To replace the heat from 1 m³ of gas using a heat pump, you need:

$$\text{Electricity (Wh)} = \frac{\text{Gas (m}^3) \times 9769 \text{(Wh/m}^3)}{4}$$


### <a id='toc1_5_10_'></a>[Task 9: Total electricity usage per household](#toc0_)

Compute the total electricity usage for each household, assuming modern heat pumps replace gas heating. To do this, **add the hourly electricity equivalent of gas to the original hourly electricity usage**. Do **not** overwrite the original electricity column. Instead, **create a new DataFrame**  `df_combined` with the following columns:
- `datetime`: the timestamp of each observation
- `date`: the date of the observation
- `combined_Wh_1`: combined usage for household 1
- `combined_Wh_2`: combined usage for household 2

Set the index of the DataFrame to the `datetime` column.

> **Recall:** you can create the DataFrame using `pd.DataFrame`, passing a dictionary containing the columns and their definition.

````python
# Write your code in the cells between the marker lines

#// BEGIN_TODO [GA_9] (4 points)

gas_to_elec = 9769 / 4 


df_household_1['elec_from_gas'] = df_household_1['gas'] * gas_to_elec
df_household_2['elec_from_gas'] = df_household_2['gas'] * gas_to_elec


combined_Wh_1 = df_household_1['electricity'] + df_household_1['elec_from_gas']
combined_Wh_2 = df_household_2['electricity'] + df_household_2['elec_from_gas']


df_combined = pd.DataFrame({'datetime': df_household_1['datetime'], 'date': df_household_1['date'], 
                            'combined_Wh_1': combined_Wh_1, 'combined_Wh_2': combined_Wh_2})


df_combined = df_combined.set_index('datetime')
````

### <a id='toc1_7_1_'></a>[Effective temperature](#toc0_)

Typically, gas consumption is higher during winter, mainly due to cold weather. However, heating demand is influenced not only by the air temperature but also by the wind. Wind increases the rate at which a house loses heat, raising heating requirements even when the measured temperature is not very low. By combining temperature and wind into one measure, we can better understand heating demands.

The **effective temperature** combines both temperature and wind speed:

$$T_{\text{effective}} = T_{\text{hour}} - \frac{2}{3} v_{\text{hour}}$$
where $T_{\text{hour}}$ is the average temperature in that hour ($^\circ \mathrm{C}$) and $v_{\text{hour}}$ is the average wind speed in that hour (m/s). The factor $\frac{2}{3}$ is used to account for wind in estimating effective temperature for heating demand calculations. It’s an empirical approximation, commonly applied in Dutch energy statistics.

There are many questions you can investigate using effective temperature instead of (normal, not corrected for wind) temperature. You can also e.g. compare a regression model using effective temperature to the model using (normal) temperature.

You can calculate the effective temperature using the formula above for each household and add it as a **new column** called `effective_temp`.

### <a id='toc1_7_2_'></a>[Degree days](#toc0_)

You may also use the concept of **degree days** to better understand heating demand.

Heating is typically switched on when the daily average outdoor temperature is below a threshold (commonly around 14 °C) and switched off otherwise. **Degree days** quantify how much heating is needed by measuring how far the temperature falls below this threshold. If the temperature is above the threshold, no heating is required, whether the daily average is 20 °C or 25 °C.

For a given day:

$$\text{degree days} = \max\{0, T_{\text{threshold}} - T_\text{effective}\},$$

where $T_{\text{threshold}}$ is the chosen threshold temparature and $T_\text{mean}$ is the daily average effective temperature.

Degree days provide a simple way to connect temperature to heating demand: larger values indicate more heating required, smaller or zero values indicate little or no heating. You can compute degree days for each day in the dataset and use them as an additional feature in your analysis to better understand energy consumption patterns.

````python
df_household_2_lm = df_household_2.copy()

weekDict = {
    'Monday' : 0,
    'Tuesday' : 0,
    'Wednesday' : 0,
    'Thursday' : 0,
    'Friday' : 0,
    'Saturday' : 1,
    'Sunday' : 1
}

df_household_2_lm['day of week (enum)'] = df_household_2_lm['day of week'].map(weekDict)

df_household_2_lm.describe()
````

````python
df_household_2_lm.describe()
````

````python
X = df_household_2_lm[['temperature', 'day of week (enum)', 'hour']]
y = df_household_2_lm['gas']
````

````python
lm_household_2 = LinearRegression()
lm_household_2.fit(X, y)
````

Now we assess how well the model manages to predict hourly gas consumption .

````python
y_predicted = lm_household_2.predict(X)
````

Add predicted gas values to the Dataframe

````python
df_household_2_lm['gas_pred'] = y_predicted
````

Now let us visualize the prediction error. Since we have 3 independent variables, it is impossible to demonstrate simply with a plot. We will use a heatmap, and we will not use the day of the week parameter.

````python

# --- INPUT ---
# DataFrame must contain:
# 'temperature', 'hour', 'gas', 'gas_pred'
df = df_household_2_lm.copy()

# --- 1. Bin temperature ---
# Adjust number of bins if needed (12–16 usually works well)
n_bins = 7
df['temp_bin'] = pd.cut(df['temperature'], bins=n_bins)

df['hour'] = pd.cut(df['hour'], bins=12)

# --- 2. Pivot tables: mean actual & mean predicted ---
pivot_actual = df.pivot_table(
    index='temp_bin',
    columns='hour',
    values='gas',
    aggfunc='mean'
)

pivot_pred = df.pivot_table(
    index='temp_bin',
    columns='hour',
    values='gas_pred',
    aggfunc='mean'
)

# --- 3. Prediction error ---
pivot_error = pivot_pred - pivot_actual

# --- 4. Plot heatmap ---
plt.figure(figsize=(7, 5))

# Symmetric color scale around zero (important!)
vmax = np.nanmax(np.abs(pivot_error.values))

im = plt.imshow(
    pivot_error.values,
    aspect='auto',
    origin='lower',
    cmap='coolwarm',
    vmin=-vmax,
    vmax=vmax
)

plt.colorbar(im, label='Prediction error (predicted − actual)')

# --- 5. Axes labels & ticks ---
plt.xticks(
    ticks=range(12),
    labels=range(12)
)

plt.yticks(
    ticks=range(len(pivot_error.index)),
    labels=[f"{b.left:.1f}–{b.right:.1f}" for b in pivot_error.index]
)

plt.xlabel('Hour of day')
plt.ylabel('Temperature bin (°C)')
plt.title('Household 2: Prediction error by temperature and hour', fontsize=16)

plt.tight_layout()
plt.show()
````

````python
lm_household_2.score(X, y)
````

Now we visialize precision of the model

````python
residuals = y - y_predicted

residuals = residuals / residuals.std(ddof=1)

df_household_2_lm['residuals'] = residuals

ax = df_household_2_lm.plot(kind='scatter', x='temperature', y='residuals', color='b', figsize=(7, 5))
plt.axhline()
ax.set_xlabel('Temperature')
ax.set_ylabel('Normalized residuals')
ax.set_title('Normalized residuals for temperature for Household 2', fontsize=16);
````

In this plot we can see that the residuals plot reveals clear structure rather than random variation, indicating violations of the linearity assumption. Residuals depend systematically on outdoor temperature and exhibit changing variance and a lower bound at zero consumption. These patterns suggest nonlinear heating behavior and interaction effects with time-related variables that are not captured by the linear model.

### Conclusion for household 1

We can conclude that hourly gas consumption can be predicted with moderate accuracy using outdoor temperature and time-related variables, with the linear regression model explaining approximately 62% of the variance. While the model captures the main temperature and daily usage patterns, residual analysis reveals systematic deviations, indicating nonlinear effects that are not fully captured by a simple linear specification.
